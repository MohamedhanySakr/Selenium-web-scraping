# Selenium in webscraping
### Selenium is a tool to automate browsers. It's primarily used for testing but is also very useful for web scraping. You can check out the unofficial documentation here.

```python
# install the selenium library (  pip install selenium  )
# You have to download a Chromedriver that is compatable with your chrome version
# https://chromedriver.chromium.org/downloads
# put it under C:/program Files (x86)
# now you are ready
```

```python
# Import the needed libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# path of the chromedriver
# driver.get to get the web page you need
PATH = 'C:/Program Files (x86)/chromedriver.exe'
driver = webdriver.Chrome(PATH)
driver.get("https://covid19.who.int/table")
```

```python
# driver.quit() to close the page 
# print(driver.page_source)
```

```python
# we can find our elements using many ways 
# driver.find_element_by_class_name
# driver.find_element_by_id
# driver.find_element_by_xpath
# driver.find_element_by_tag_name
# etc...
```

```python
# find elements by id 
# the first id in the page is ___gatsby
try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "___gatsby"))
    )
    print(main.text)
finally:
    driver.quit()
```

```python
# or you can do this 
first = driver.find_element_by_id("___gatsby")
print(first.text)
```

```python
# to click on a link or move to another page 
Link = driver.find_element_by_link_text('Overview')
Link.click()
```

```python
# find element by xpath is good if you want a specific data
The_Path = driver.find_element_by_xpath('.//*[@id="gatsby-focus-wrapper"]/div/div[4]')
```

```python
print(The_Path.text)
```

```python
# find elemtents by xpath is differ from find element by xpath as your data is a list  
# we can append the data in another variable
My_list = []
Thetitles = driver.find_elements_by_xpath('.//*[@id="gatsby-focus-wrapper"]/div/div[4]/div/div[2]/div/div')
for job in Thetitles:
    My_list.append(job.get_attribute('href'))
```
